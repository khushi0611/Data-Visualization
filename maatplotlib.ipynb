```python
import matplotlib.pyplot as plt
import numpy as np

days = list(range(1, 31))
temperature = np.random.randint(20, 50, size=30)

# Create a line plot
plt.plot(days, temperature, marker='*')

plt.xlabel("Days")
plt.ylabel("Temperature (C)")
plt.title("Temperature Over 30 Days")

plt.show()
```

```python
import matplotlib.pyplot as plt
days=["mon",'tue','wed','thru','fri','sat','sun'] 
temperature = np.random.randint(20, 50, size=7)

plt.figure(figsize=(8,6))
plt.bar(days,temperature,color=['blue','green','brown'],width=0.4,edgecolor='y',linewidth=2,linestyle=':',alpha=0.8,label='Temperature')

plt.xlabel("Days",fontsize=14)
plt.ylabel("Temperature (C)",fontsize=14)
plt.title("Temperature Over a  week ",fontsize=16)
plt.xticks(rotation=45) #rotte x-axis label if they are long

plt.tight_layout()
plt.legend()
#showing plot
plt.show()
```
