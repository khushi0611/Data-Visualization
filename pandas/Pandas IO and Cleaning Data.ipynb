```python
import pandas as pd

import numpy as np

#Sample Series

s=pd.Series(['a', 'b', 'a', 'c', 'b', 'a', np.nan, 'c', 'c', 'c'])

#Basic value counts

print("Basic value counts:")

print(s.value_counts())

#Do not sort by count (order of appearance)

print("\nValue counts without sorting:")

print(s.value_counts(sort=False))

# Sort results in ascending order

print("\nValue counts sorted in ascending order:")

print(s.value_counts(ascending=True))

#Include NaN values in the counts

print("\nValue counts including NaN:")

print(s.value_counts(dropna=False))
```

### nsamllest()

**Syntax**: DataFrame.nsmallest(n, columns, keep='first')

  - **n**: An integer representing the number of largest values you want to retrieve from the
           Series or DataFrame.

  - **columns**: In the DataFrame version, you specify the column(s) to consider when finding
                the largest values. This can be a single column name or a list of column names.

  - **keep**: (Optional) Determines how to handle duplicates if there are multiple values with
              the same highest value. It can be one of the following:

              'first': Keep the first occurrence of duplicates (default).
              
              'last': Keep the last occurrence of duplicates.

```python
import pandas as pd
df=pd.read_csv('Book1.csv')

df.nsmallest(4,'basepay')
```

## Summarizing or grouping data
### 1. group by()
 #### **Grouping by one column**

```python
import pandas as pd
csv_file_path="Book1.csv"
df=pd.read_csv(csv_file_path)
print(df[['basepay','rating']].groupby(df['gender']).min())

print(df[['basepay']].groupby(df['designation']).sum())
```

 #### **Grouping by multiple column**

```python
import numpy as np 
import pandas as pd
data={
    'Date': ['2024-01-01','2024-01-02','2024-01-03','2024-01-04','2024-01-05'],
    'Category':['a','b','a','b','a'],
    'Sales':[100.234,112.343,345.987,456.086,567.980],
    'Quantity':[10,20,30,40,50]
}
df=pd.DataFrame(data)
pivot_table=pd.pivot_table(df,values=['Sales'],index=['Date'],columns=['Category'],aggfunc={'Sales':'sum'},fill_value=0,dropna=False,margins=True, margins_name='Total')
np.round(pivot_table,0)
```
