**Deletion**

1. From Specific Position
  * pop() - to delete element from specific position.
     - Syntax - listname.pop(index)
     - if index not specified , then last element get deleted.



```python
# to insert all the element of any other sequence datatype into a given list at specified position as seperate elements.

# python do not provide predefined library function for this task because we have to define position for every element. default value is not want by the user everytime
```

```python
firstlist=[20,30,40,20,30,50,70,50]
print("list is : ", firstlist)

firstlist.pop()
print("list is : ", firstlist)

firstlist.pop(4)
print("list is : ", firstlist)

firstlist.pop(2)
```

2. Removing an element
  - remove() - to remove specific element from the position of its first occurrence from left most element into the list
  - Syntax - listname.remove(element)
  - if element is not present then exception is generated automatically.

```python
list2=[20,30,40,20,30,50,70,50]
print("list is : ", list2)

list2.remove(50)
print("list is : ", list2)

list2.remove(20)
```

3. delete all elements
  - clear()- delete all elements
  - Syntax- listname.clear()

4. deleting elements in fiven range
  - del()
    - syntax - del listname[start : end : step]

```python
list2=[20,30,40,20,30,50,70,50]
print("list is : ", list2)

del list2[2:6:2]
print("list is : ", list2)
```

```python
# Display element of tuple in forward direction using backward index.
print("Element of tuple in forward direction using backward index-")
for index in range(-len(tuple),0):
  print(tuple[index],end=" ")

# Display element of tuple in backward direction using farward index.
print('\nElement of tuple in backward direction using farward index-')
for index in range(len(tuple)-1,-1,-1):
  print(tuple[index],end=" ")
```

# Methods of dictionary
* pop()-
    - removes and returns the specified element from the dictionary.
    - dict_name.pop(key)
* popitem()-
    - removes and returns the last element from the dictionary.
    - dict.popitem(last = True)
    - dict.popitem(last = False)
* keys() -
    - To return a list of keys of dictionary
    - listname = dict_name.keys()  OR

    - print(dict_name.keys())
* values()
    - To return the value of elements in a dictionary
    - listname = dict_name.values()
    -   OR
    -   print(dict_name.values())
* items()
    - To return a list of items from dictionary(each item is a tuple which contains key,value of each element )
    - listVar = dict_name.items
* get()
    - to return value of element with specific key
    - variable = dictname.get(key, error_msg)
* clear()
    - to delete all the elements from the dictionary
    - dic_name.clear()
    - the dictionary will be deleted but the object will not be deleted
* len()
    - to return total number of elements
    - variable = len(dic_mame)
* update()
    - to insert/update element of any other dictionary
    - dic1_name.update(dic2_name)

```python
# Dictionaries are often used to store many pieces of information.
# Dictionaries are created by enclosing key-value pairs within braces or curly brackets { and }
person1 = {
    'name': 'John Doe',
    'sex': 'Male',
    'age': 32,
    'married': True
}
print(person1)

# Dictionaries can also be created using the dict function.

person2 = dict(name='Jane Judy', sex='Female', age=28, married=False)
print(person2)
```
