```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt


data_path = 'Book1.csv'
df = pd.read_csv(data_path)

rating = df['rating']  
designation = df['designation'] 

# bar plot
plt.figure(figsize=(10,7))
plt.subplot(2,2,1)
plt.bar(designation, rating, width=0.4, linewidth=2, linestyle=':', alpha=0.8)
plt.xlabel("Designation", fontsize=14)
plt.ylabel("Rating", fontsize=14)
plt.title("Rating according to Gender Ratio", fontsize=16)

gender = df['gender']  
plt.subplot(2,2,2)
f_count= df.groupby('gender')['gender'].count()
# plt.pie(gender)
f_count.plot(kind = 'pie')
plt.title("", fontsize=16)
plt.title("Gender Ratio ")

 
plt.subplot(2,2,3)
basepay = df['basepay']  
designation = df['designation'] 

# line plot
plt.bar(basepay, designation)


plt.xlabel("basepay", fontsize=14)
plt.ylabel("designation", fontsize=14)






plt.xticks(rotation=45)


# plt.tight_layout()


plt.show()
```

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt


data_path = 'Book1.csv'
df = pd.read_csv(data_path)

gender = df['gender']  
designation = df['designation'] 

# line plot
plt.plot(gender, designation)


plt.xlabel("gender", fontsize=14)
plt.ylabel("designation", fontsize=14)
plt.title("Rating according to Gender Ratio", fontsize=16)


plt.xticks(rotation=45)


plt.tight_layout()


plt.show()
```
