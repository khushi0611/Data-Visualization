```python
file = open("First file.txt",'w')
file.write("Hello World...")
print("Written into file successfully...")
file.close()
```

```python
# with open
with open ("First file.txt",'a') as file:
    file.write("Welcome to python class")

with open ("First file.txt",'r') as file:
    read = file.read()
print(read)
```

```python
# tell() and seek()
with open ("First file.txt",'r') as file:
    print(file.tell())
    file.seek(4)
    data = file.read(10)
    print(file.tell())
print(data)
search = input("enter the text which you want to search : ")
with open("First file.txt",'r') as file:
    for line in file:
        if search.lower() in line.lower():
            print(line.strip())
print(f"Search for'{search}' completed")
```

```python
# pickel method
import pickle
file = open("file.dat",'ab')
li = [20,30,40,50,60]
pickle.dump(li,file)
print("done")

file = open("file.dat",'rb')
data = pickle.load(file)
print(data)
file.close()
```

```python
# write a function in python to count uppercase character in a text file "ABC.txt"
count=0
with open('ABC.txt','r') as file:
  content = file.read()
  for char in content:
    if char.isupper():
      count=count+1
print(count)
```
